```python
# Access your secret keys via
from google.colab import userdata
# The name of your secret must match `OPENAI_API_KEY`
OPENAI_API_KEY = userdata.get('OPENAI_API_KEY')

# Import OpenAI API and set up the key
from openai import OpenAI
client = OpenAI(api_key=(OPENAI_API_KEY))
```

```python
# Assignment 3 – GPT: The Worst Mathematician Ever

n = int(input("Enter base number (n): "))
i = int(input("Enter number of iterations (i): "))

value = n
errors = 0

for step in range(i):
    prompt = f"Multiply {value} by itself. Only return the number."

    response = client.responses.create(
        model="gpt-4.1-mini",
        temperature=0,
        max_output_tokens=20,
        input=prompt
    )

    try:
        gpt_result = int(response.output_text.strip())
    except:
        gpt_result = None

    correct = value * value

    print(f"\nStep {step + 1}")
    print("GPT:", gpt_result)
    print("Correct:", correct)

    if gpt_result != correct:
        errors += 1
        print("Ugh… I messed up again. Math is hard.")
        value = correct
    else:
        value = gpt_result

print("\nFinal value:", value)
print("Total errors:", errors)
```
